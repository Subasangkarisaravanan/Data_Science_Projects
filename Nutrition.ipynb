```python
import pycountry
import pandas as pd
```

```python
special_cases = {
    'GLOBAL': 'Global',
    'WB_LMI': 'Low & Middle Income',
    'WB_HI': 'High Income',
    'WB_LI': 'Low Income',
    'WB_UMI': 'Upper Middle Income',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'SEAR': 'South-East Asia Region',
    'WPR': 'Western Pacific Region',
    'AMR': 'Americas Region'
}
```

```python
def convert_country_code(code):
    if pd.isna(code):
        return code

    # Handle special region / income codes
    if code in special_cases:
        return special_cases[code]

    try:
        country = pycountry.countries.get(alpha_3=code)
        return country.name if country else code
    except:
        return code
```

```python
df_obesity['Country'] = df_obesity['Country'].apply(convert_country_code)
df_obesity.head()
```

```python
df_malnutrition['Country'] = df_malnutrition['Country'].apply(convert_country_code)
df_malnutrition.head()
```

```python
df_obesity['CI_Width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']
df_obesity.head()
```

```python
df_malnutrition['CI_Width'] = df_malnutrition['UpperBound'] - df_malnutrition['LowerBound']
df_malnutrition.head()
```

```python
df_obesity['obesity_level'] = df_obesity['Mean_Estimate'].apply(
    lambda x: 'High' if x >= 30
    else 'Moderate' if 25 <= x <= 29.9
    else 'Low'
)
df_obesity.head()
```

```python
df_malnutrition['malnutrition_level'] = df_malnutrition['Mean_Estimate'].apply(
    lambda x: 'High' if x >= 20
    else 'Moderate' if 10 <= x <= 19.9
    else 'Low'
)
df_malnutrition.head()
```

```python
df_malnutrition['age_group'] = (
    df_malnutrition['age_group']
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        'adults': 'Adult',
        'children': 'Child/Adolescent'
    })
)
```

```python
import sqlite3

# Create SQLite database (auto-created if not exists)
conn = sqlite3.connect("nutrition_paradox.db")

# Create cursor object
cursor = conn.cursor()

print("Database connected successfully")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS obesity (
    Year INTEGER,
    Gender TEXT,
    Mean_Estimate REAL,
    LowerBound REAL,
    UpperBound REAL,
    Age_Group TEXT,
    Country TEXT,
    Region TEXT,
    CI_Width REAL,
    Obesity_Level TEXT
)
""")
cursor.execute("""
CREATE TABLE IF NOT EXISTS malnutrition (
    Year INTEGER,
    Gender TEXT,
    Mean_Estimate REAL,
    LowerBound REAL,
    UpperBound REAL,
    Age_Group TEXT,
    Country TEXT,
    Region TEXT,
    CI_Width REAL,
    Malnutrition_Level TEXT
)
""")

print("Tables created successfully")
```

```python
df_obesity.rename(columns={
    "age_group": "Age_Group",
    "obesity_level": "Obesity_Level"
}, inplace=True)

df_malnutrition.rename(columns={
    "age_group": "Age_Group",
    "malnutrition_level": "Malnutrition_Level"
}, inplace=True)
```

```python
df_obesity = df_obesity[
    ["Year", "Gender", "Mean_Estimate", "LowerBound", "UpperBound",
     "Age_Group", "Country", "Region", "CI_Width", "Obesity_Level"]
]

df_malnutrition = df_malnutrition[
    ["Year", "Gender", "Mean_Estimate", "LowerBound", "UpperBound",
     "Age_Group", "Country", "Region", "CI_Width", "Malnutrition_Level"]
]
```

```python
for _, row in df_obesity.iterrows():
    cursor.execute("""
        INSERT INTO obesity VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        int(row["Year"]),
        row["Gender"],
        float(row["Mean_Estimate"]),
        float(row["LowerBound"]),
        float(row["UpperBound"]),
        row["Age_Group"],
        row["Country"],
        row["Region"],
        float(row["CI_Width"]),
        row["Obesity_Level"]
    ))
```

```python
for _, row in df_malnutrition.iterrows():
    cursor.execute("""
        INSERT INTO malnutrition VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        int(row["Year"]),
        row["Gender"],
        float(row["Mean_Estimate"]),
        float(row["LowerBound"]),
        float(row["UpperBound"]),
        row["Age_Group"],
        row["Country"],
        row["Region"],
        float(row["CI_Width"]),
        row["Malnutrition_Level"]
    ))
```

```python
conn = sqlite3.connect("nutrition_paradox.db")

query = """
SELECT o.Country
FROM obesity o
JOIN malnutrition m
ON o.Country = m.Country
GROUP BY o.Country
HAVING MAX(o.Mean_Estimate) > MIN(o.Mean_Estimate)
   AND MAX(m.Mean_Estimate) < MIN(m.Mean_Estimate);
"""

df = pd.read_sql(query, conn)
print(df)

conn.close()
```
